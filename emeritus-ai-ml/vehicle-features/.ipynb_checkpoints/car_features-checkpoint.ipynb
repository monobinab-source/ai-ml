#### Extract car age from Year column by simple arithmetic calculation

#### Extract the numeric attribute from string value
For cylinders the data is in the format of e.g. "6 cylinders". From this string we need to extract the numeric value of 6 by doing simple string processing for this column using regex.

```python
import re
print(vehicles_1['cylinders'].head(10))

# Convert column to string (ensures compatibility)
vehicles_1['cylinders'] = vehicles_1['cylinders'].astype(str)

# Create an empty list to store extracted numbers
cylinder_numbers = []

# Iterate over each row in the 'cylinders' column
for value in vehicles_1['cylinders']:
    match = re.search(r'\d+', value)  # Find digits in the string
    if match:
        cylinder_numbers.append(int(match.group()))  # Convert to integer
    else:
        cylinder_numbers.append(None)  # If no match, append None (or np.nan)

# Assign the extracted values back to the dataframe column
vehicles_1['cylinders'] = cylinder_numbers

# Print the first 10 rows to verify
print(vehicles_1['cylinders'].head(10))  # Output: 12
vehicles_1.head()
vehicles_1.info()
```

#### Handle categorical features having high cardinality in different approach
One-hot encoding for columns with a large number of unique values (over 50 different values), like manufacturer and model, can lead to a massive increase in the number of features, which can make your dataset inefficient to process and potentially cause overfitting. 

#### Target encoding with mean price for the category
This approach groups the same manufacturer cars into one group and calculate the mean price and populate manufacturer column with the mean price value.
Similarly model, state and region are done.

```python
# Feature Engineering

# Target encoding for manufacturer
#manufacturer_mean_price = vehicles_filtered_100K.groupby('manufacturer')['price'].mean()
manufacturer_mean_price = round(vehicles_encoded.groupby('manufacturer')['price_scaled'].mean(), -2)
vehicles_encoded['manufacturer_encoded'] = vehicles_encoded['manufacturer'].map(manufacturer_mean_price)

# Target encoding for model
#model_mean_price = vehicles_filtered_100K.groupby('model')['price'].mean()
model_mean_price = round(vehicles_encoded.groupby('model')['price_scaled'].mean(), -2)
vehicles_encoded['model_encoded'] = vehicles_encoded['model'].map(model_mean_price)

state_mean_price = round(vehicles_encoded.groupby('state')['price_scaled'].mean(), -2)
vehicles_encoded['state_encoded'] = vehicles_encoded['state'].map(state_mean_price)

region_mean_price = round(vehicles_encoded.groupby('region')['price_scaled'].mean(), -2)
vehicles_encoded['region_encoded'] = vehicles_encoded['region'].map(state_mean_price)

# print(vehicles_filtered_100K['model_encoded'].unique().size)
# print(vehicles_filtered_100K['manufacturer_encoded'].unique().size)
# print(manufacturer_model_mean_price)
# print(vehicles_filtered_100K['manufacturer_model_encoded'].unique().size)
# print(vehicles_filtered_100K['manufacturer_model_encoded'])

# vehicles_filtered_100K.info()
# vehicles_filtered_100K.head(25)

print(vehicles_encoded.info())
vehicles_encoded.head(25)
```

## 4. Modeling
Model Selection:
Use regression models to predict price (e.g., Linear Regression, Random Forest, or Gradient Boosting).
Start with simple models and progress to more complex ones if necessary.
Evaluate Models:
Use metrics like Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and R to assess performance.

### Trying to identify important features: Apply Regularization 
#### LASSO regularization (L1):
Lasso (Least Absolute Shrinkage and Selection Operator) adds an L1 penalty to the loss function in linear regression:
Key Properties of L1 Regularization (Lasso)
1. Feature Selection: Can shrink some coefficients exactly to zero, effectively removing irrelevant features.
2. Sparse Models: Helps in reducing model complexity by keeping only the most important features.
3. Useful for High-Dimensional Data: Works well when there are many features but only a few are truly important.


### Grid Search CV (Cross-Validation) 

It is a hyperparameter tuning method that helps find the best combination of hyperparameters for a given model.

How It Works:
1. You define a range of hyperparameter values you want to test.
2. Grid Search tries every possible combination of these hyperparameters.
3. It uses cross-validation to evaluate the performance of each combination.
4. It selects the best hyperparameters based on performance metrics (like accuracy, RMSE, etc.).

Example Use Case:
Let's say you're training a Lasso regression model, and you want to find the best value of alpha. You can use GridSearchCV to automatically test different values of alpha and select the best one.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

# Define the model
lasso = Lasso()

# Define hyperparameter grid
param_grid = {'alpha': [0.1, 0.01, 0.001, 0.0001]}

# Set up Grid Search with Cross-Validation
grid_search = GridSearchCV(lasso, param_grid, cv=5, scoring='r2')

# Fit to the data
grid_search.fit(X_train, y_train)

# Best hyperparameter
print("Best alpha:", grid_search.best_params_)

# Best model
best_lasso = grid_search.best_estimator_
```

```python
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import Ridge

def get_training_dev_MSE_for_model_with_alpha(alpha, X_train, y_train, X_test, y_test):
    # Define a Ridge regression model with polynomial features and scaling
    scaled_ridge_model = Pipeline([
        ('poly_features', PolynomialFeatures(degree=3, include_bias=False)),  # Fixed typo
        ('scaler', StandardScaler()),
        ('ridge_regression', Ridge(alpha=alpha))
    ])
    
    # Fit model on training data
    scaled_ridge_model.fit(X_train, y_train)
    
    # Compute training and validation MSE
    training_error = mean_squared_error(y_train, scaled_ridge_model.predict(X_train))
    validation_error = mean_squared_error(y_test, scaled_ridge_model.predict(X_test))
    
    return alpha, training_error, validation_error

# Define hyperparameter grid for Ridge regression
parameters_to_try = {'ridge_regression__alpha': [0.01, 0.1, 1, 10, 100]}

# Cross-validation strategy
cv_strategy = KFold(n_splits=5, shuffle=True, random_state=42)  # Use KFold instead of incorrect cv list

# Initialize GridSearchCV
model_finder = GridSearchCV(
    estimator=Pipeline([
        ('poly_features', PolynomialFeatures(degree=3, include_bias=False)),
        ('scaler', StandardScaler()),
        ('ridge_regression', Ridge())
    ]),
    param_grid=parameters_to_try,
    scoring="neg_mean_squared_error",
    cv=cv_strategy
)

# Fit GridSearchCV on training data
model_finder.fit(X_train, y_train)

# Get best alpha and corresponding error
best_alpha = model_finder.best_params_['ridge_regression__alpha']
best_mse = -model_finder.best_score_

print(f"Best Alpha: {best_alpha}")
print(f"Best Cross-Validation MSE: {best_mse}")
```
